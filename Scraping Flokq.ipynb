```python
import requests
```

```python
url = 'https://prod.strapicms.flokq.com/rooms/rooms-and-units'

#parameter
parameter = {
'workflow.status': 'published',
'apartment.ZoneRelation.province.Name': 'Jakarta',
'exclude': 'Kost',
'Status': 'Available',
'_start': 30,
'_limit': 1790,
'_sort': 'Ranking:DESC',
'language.language': 'id',
}

#API requests
r = requests.get(url, params= parameter).json()
```

```python
#scraping
apartments = r['rooms']
nama, harga, luas, kt, km, lantai, lokasi, kota, furnished, min_stay, landmark, landmark_2, fasilitas, maintenance, latitude, longitude = [],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]

for data in apartments:
    nama.append(data['apartment']['name'])
    harga.append(data['TotalPrice'])
    luas.append(data['FloorArea'])
    kt.append(data['NumOfBedrooms'])
    km.append(data['NumOfBathroom'])
    lantai.append(data['Floor'])
    lokasi.append(data['apartment']['Neighborhood'])
    kota.append(data['apartment']['zone'])
    furnished.append(data['FurnishingType'])
    min_stay.append(data['PaymentOption'])
    landmark.append(data['apartment']['ComfortConvenienceDescription'])
    landmark_2.append(data['apartment']['NeighborhoodDescription'])
    fasilitas.append(data['apartment']['FacilitiesDescription'])
    maintenance.append(data['apartment']['Maintenance'])
    latitude.append(data['apartment']['lat'])
    longitude.append(data['apartment']['long'])
    
    
    
```

```python
#Save Data

listcols = ['nama', 'harga', 'luas', 'kt', 'km', 'lantai', 'lokasi', 'kota', 'furnished', 'min_stay', 'landmark', 'landmark_2', 'fasilitas', 'maintenance', 'latitude', 'longitude']
dict_data = dict(zip(listcols,(nama, harga, luas, kt, km, lantai, lokasi, kota, furnished, min_stay, landmark, landmark_2, fasilitas, maintenance, latitude, longitude)))
```

```python
import pandas as pd
```
